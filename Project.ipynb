**<h1>3 Assessment of the Impact of Experience Levels</h1>**

In this section, I examine how experience levels affect salary in the data profession. My goal is to understand the relationship between an individual's experience and earning potential, which is an important factor for anybody considering a career in this dynamic area.

**3.1 Understanding Experience Levels**

Professionals are often divided into four levels of experience: entry-level, mid-level, senior, and executive. Each of these categories reflects a crucial stage in a professional's career, and is generally distinguished by various skill sets, responsibilities, and, as a result, salary ranges.

**3.2 Analysis**


```python
# Grouping by experience_level
salary_by_experience = df.groupby(['experience_level'])['salary_in_usd'].mean()

# Sorting the results for better visualization
salary_by_experience = salary_by_experience.sort_values(ascending=False)
print(salary_by_experience)
```

```python
# Plotting the average salary by experience level
salary_by_experience.plot(kind='bar', figsize=(12, 8), title='Average Salary by Experience level (USD)')
```

```python
# Grouping by job_category and experience_level
salary_by_role_experience = df.groupby(['job_category', 'experience_level'])['salary_in_usd'].mean()

# Sorting the results for better visualization
salary_by_role_experience = salary_by_role_experience.sort_values(ascending=False)
salary_by_role_experience.head(5)
```

```python
# Plotting the average salary by role and experience level
salary_by_role_experience.unstack().plot(kind='bar', figsize=(10, 6), title='Average Salary by Role and Experience Level (USD)')
```

Looking at the bar graph, I may make numerous conclusions and observations for my analysis:

1. **Progression with Experience:**
* Salary levels are clearly rising as experience levels rise across all job categories. This is to be expected, and it is consistent with the widespread idea that greater experience often correlates with better pay. For example, the 'Machine Learning and AI' category displays a significant variation in income from Entry-level to Executive, showing a notable earnings development curve.

2. **Field Specific Trends:**
* The graph also shows industry-specific trends, such as 'Machine Learning and AI' and 'Data Science and Research' commanding greater compensation across all experience levels. This confirms my belief that certain specialisations are more profitable than others due to their complexity and strong demand for advanced talents.

3. **Senior-Level Saturation**
* In certain professions, such as 'Data Management and Strategy,' the compensation rise from Senior to Executive levels appears to be less prominent than in others. According to my interpretation, this might indicate saturation at the senior level, where advancing into executive positions could not be as closely linked to pay increases as it is in other categories.

**<h1>4 Correlation between Employment Type and Work Setting</h1>**

In this part, I will look at the link between employment type and work setting to see how these variables may affect compensation levels in the data profession. This research is essential for understanding the current work environment and its influence on compensation.

**4.1 Defining Employment Types and Work Settings**

* **Employment Type:** This covers Full-time, Part-time, Contract, and Freelance positions. Each of them has unique qualities in terms of job security, perks, and work hours required.
* **Work Setting:** This refers to where and how professionals carry out their duties, such as through remote, in-person, or hybrid work arrangements.

**4.2 Analysis**

```python
# Group the data by employment type and work setting
# Calculate the average salary for each group
grouped_data = df.groupby(['employment_type', 'work_setting'])['salary_in_usd'].mean().reset_index()
grouped_data.head(10)
```

```python
# Pivot the data to prepare for plotting
pivot_data = grouped_data.pivot(index='employment_type', columns='work_setting', values='salary_in_usd')

# Plotting the average salary by employment type and work setting
ax = pivot_data.plot(kind='bar', figsize=(10, 6), title='Average Salary by Employment Type and Work Setting (USD)')

# Setting the labels and legend
ax.set_xlabel('Employment Type')
ax.set_ylabel('Average Salary in USD')
ax.legend(title='Work Setting')

# Show the plot
plt.show()
```


Analyzing the bar graph that presents the average salary by Employment Type and Work Setting, I can draw several conclusions and insights:

* **In-person Work:** According to the graph, full-time employment have the greatest average salary when the work environment is in-person, implying that professions requiring physical presence may offer more competitive remuneration. 

* **Remote Work:** Remote work environments offer large compensation for full-time employment, although not as high as in-person settings, contrary to what may be expected given the current trend towards remote work. This might imply that, while remote employment allows for greater freedom, there may be positions in the data industry where being on-site is still financially desirable.

* **Hybrid Work:** Full-time earnings in hybrid settings are lower than remote positions but higher than part-time roles, indicating a midway ground. This might be because hybrid arrangements provide a good blend of flexibility and office presence.

* **Contract and Freelance Work:** Contract employment has a high average wage across all work contexts, with remote contract work being fairly competitive but not as high as full-time in-person positions. Because of the unpredictability and fluctuation of freelancing labour, average pay are often the lowest.

* **Part-time Work:** Part-time jobs, regardless of work environment, have the second lowest average wages, which is to be anticipated given the less hours worked. Part-time in-person positions pay more on average than remote tasks, which may indicate a desire for in-person cooperation in part-time settings or the nature of part-time roles that necessitate physical presence.

**<h1>5 Salary Trends by company Size</h1>**

In this part, I'll look at how the size of a firm affects data professionals' compensation. Understanding this relationship is critical because it will impact my compensation and business culture expectations once I enter the workforce.

**5.1 Definition of Company Sizes**


Small (S), medium (M), and large (L) are the most common company sizes. Because of the resources available, the breadth of projects, and the opportunity for progression within the organisation, these categories can have a significant influence on wage scales.

**5.2 Analysis**

```python
# Group the data by company size
# Calculate the average salary for each group
average_salary_by_company_size = df.groupby('company_size')['salary_in_usd'].mean()
average_salary_by_company_size.head()
```

```python
# Plotting the average salary by company size
average_salary_by_company_size.plot(kind='barh', title='Average Salary by Company Size (USD)')
```

When I explore the given wage data by business size, I see a trend that slightly defies commonly held beliefs about firm size and income.

This excites me because it implies a sweet spot between the agility of tiny businesses and the resources of huge ones. It looks to be an atmosphere filled with potential, in which my efforts may have a large influence while yet providing a respectable wage.

* **Large Companies:** Large corporations follow closely after, with an average compensation of roughly $141,097. The difference between large and medium-sized businesses is significant, and it is something I will absolutely examine when I assess the sort of corporate culture and structure I want to join.
